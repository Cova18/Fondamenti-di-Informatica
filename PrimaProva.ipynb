#### Esercizio 17
##### Build chains of trips. Two trips are consecutive in a chain if (a) they have the same VendorID, (b) the pickup location of the second trip is also the dropoff location of the first trip, (c) the pickup time of the second trip is after the dropoff time of the first trip, and (d) the pickup time of the second trip is at most 2 minutes later than the dropoff time of the first trip.

```python
df_nuovo= dataSet.sort_values(["VendorID", "tpep_pickup_datetime"]).reset_index()
df_nuovo["dropoff_time"] = df_nuovo.groupby("VendorID")["tpep_dropoff_datetime"].shift()
#nella riga precedente ho messo un group by sul vendor cosi vincolo lo shif per tipo di vendor
#perche anche se ho tutto raggruppato rischio di shiftare la riga il valore di un vendor su un altro

#sembra abbia funzionato ripeto per la location
df_nuovo["dropoff_location"] = df_nuovo.groupby("VendorID")["DOLocationID"].shift()

df_nuovo["chan"] = ((df_nuovo["PULocationID"] == df_nuovo["dropoff_location"]) & (pd.to_datetime(df_nuovo["tpep_pickup_datetime"]) > pd.to_datetime(df_nuovo["dropoff_time"])) &
                    ((pd.to_datetime(df_nuovo["tpep_pickup_datetime"]) - pd.to_datetime(df_nuovo["dropoff_time"])).dt.total_seconds() <= 120))
df_nuovo
```
