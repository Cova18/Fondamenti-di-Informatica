#### Step 4

```python
bad = dataFrame.query('payment_type.isnull() | VendorID.isnull()| passenger_count.isnull()| store_and_fwd_flag.isnull()')
dataFrame = dataFrame.drop(bad.index)
dataFrame
```

#### Step 5
Add a duration column storing how long each trip has taken (use <font color = 'red'>tpep_pickup_datetime, tpep_dropoff_datetime</font>)

```python
dataFrame['trip_duration'] = (pd.to_datetime(dataFrame['tpep_dropoff_datetime']) - pd.to_datetime(dataFrame['tpep_pickup_datetime']))#.dt.seconds
dataFrame.head()
```

#### Step 6
For each pickup location, determine how many trips have started there.

```python
count_pick_up =  dataFrame.groupby(['PULocationID']).size()
count_pick_up
```

#### Step 7
Cluster the pickup time of the day into 30-minute intervals (e.g. from 02:00 to 02:30)

```python
dataFrame['intervallo'] = pd.to_datetime(dataFrame['tpep_pickup_datetime']).dt.floor("30min")
```

#### Step 8
For each interval, determine the average number of passengers and the average fare amount.

```python
intervalli_df = dataFrame.groupby(["intervallo"]).mean(["passenger_count","fare_amount"])[["passenger_count", "fare_amount"]]
intervalli_df
```

#### Step 9
For each payment type and each interval, determine the average fare amount

```python
df_pay_int = dataFrame.groupby(["intervallo","payment_type"]).mean(["fare_amount"])[["fare_amount"]].reset_index()
df_pay_int
```

#### Step 10
For each payment type, determine the interval when the average fare amount is maximum

```python
df_pay_int.loc[df_pay_int.groupby("payment_type")["fare_amount"].idxmax()].reset_index()
```

#### Step 11
For each payment type, determine the interval when the overall ratio between the tip and the fare amounts is maximum

```python
dataFrame = dataFrame[(dataFrame["fare_amount"] != 0) & (dataFrame["fare_amount"].notna()) & (dataFrame["tip_amount"].notna())]
dataFrame["ratio"] = dataFrame["tip_amount"] / dataFrame["fare_amount"]     
df11 = dataFrame[["intervallo","payment_type","tip_amount","fare_amount","ratio"]]
#df11.groupby("payment_type")["ratio"].idxmax()
df11.loc[df11.groupby("payment_type")["ratio"].idxmax()].reset_index(drop = True)
```

#### Step 12
Find the location with the highest average fare amount

```python
tmp = dataFrame[["PULocationID", "fare_amount"]].rename(columns = {"PULocationID":"Location"})
tmp2 =  dataFrame[["DOLocationID", "fare_amount"]].rename(columns = {"DOLocationID":"Location"})
new = pd.concat([tmp, tmp2], ignore_index=True)
max_location = new.groupby("Location").mean().idxmax()
max_location
```
